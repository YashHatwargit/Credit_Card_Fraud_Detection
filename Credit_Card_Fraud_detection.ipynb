```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
# first 5 rows of the dataset
credit_card_data.head()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

This Dataset is highly unblanced

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression(max_iter=2000)
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report

# Predict on test data
X_test_prediction = model.predict(X_test)

# Print individual metrics
print("Precision:", precision_score(Y_test, X_test_prediction))
print("Recall:", recall_score(Y_test, X_test_prediction))
print("F1 Score:", f1_score(Y_test, X_test_prediction))

# Full report (includes support, accuracy, etc.)
print("\nClassification Report:\n")
print(classification_report(Y_test, X_test_prediction))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Initialize the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=2)

# Train the model
rf_model.fit(X_train, Y_train)

# Predict on the test set
rf_test_prediction = rf_model.predict(X_test)

# Accuracy
accuracy = accuracy_score(Y_test, rf_test_prediction)
print("Accuracy:", accuracy)

# Precision
precision = precision_score(Y_test, rf_test_prediction)
print("Precision:", precision)

# Recall
recall = recall_score(Y_test, rf_test_prediction)
print("Recall:", recall)

# F1 Score
f1 = f1_score(Y_test, rf_test_prediction)
print("F1 Score:", f1)

# Full Classification Report
print("\nClassification Report:\n")
print(classification_report(Y_test, rf_test_prediction))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Initialize the XGBoost model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=2)

# Train the model
xgb_model.fit(X_train, Y_train)

# Predict on the test set
xgb_test_prediction = xgb_model.predict(X_test)

# Accuracy
accuracy = accuracy_score(Y_test, xgb_test_prediction)
print("Accuracy:", accuracy)

# Precision
precision = precision_score(Y_test, xgb_test_prediction)
print("Precision:", precision)

# Recall
recall = recall_score(Y_test, xgb_test_prediction)
print("Recall:", recall)

# F1 Score
f1 = f1_score(Y_test, xgb_test_prediction)
print("F1 Score:", f1)

# Full Classification Report
print("\nClassification Report:\n")
print(classification_report(Y_test, xgb_test_prediction))
```
